```python
#To print first 10 natural numbers using function
def natural_numbers():
    for i in range(1, 11):
        print(i)

natural_numbers()
```

```python
#To calculate sum of first N natural numbers
def sum_natural(n):
    total = 0
    for i in range(1, n + 1):
        total = total + i
    print("Sum =", total)

sum_natural(10)
```

```python
#To reverse a number using function
def reverse_num(num):
    rev = 0

    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num = num // 10

    print("Reversed Number =", rev)

reverse_num(1234)
```

```python
#To count digits in a number
def count_digits(num):
    count = 0

    while num > 0:
        num = num // 10
        count = count + 1

    print("Number of Digits =", count)

count_digits(1234)
```

```python
#To check palindrome number using function
def palindrome(num):
    original = num
    rev = 0

    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num = num // 10

    if original == rev:
        print("Palindrome Number")
    else:
        print("Not a Palindrome Number")

palindrome(121)
```

```python
# To generate Fibonacci series using function

def fibonacci(n):
    a = 0
    b = 1

    print("Fibonacci Series:")

    for i in range(n):
        print(a, end=" ")
        c = a + b
        a = b
        b = c

# Number of terms
n = int(input("Enter number of terms: "))
fibonacci(n)
```

```python
# Calculator using Functions

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b

print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

choice = int(input("Enter your choice (1-4): "))

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

if choice == 1:
    print("Result =", add(num1, num2))

elif choice == 2:
    print("Result =", subtract(num1, num2))

elif choice == 3:
    print("Result =", multiply(num1, num2))

elif choice == 4:
    if num2 != 0:
        print("Result =", divide(num1, num2))
    else:
        print("Division by zero is not allowed")

else:
    print("Invalid Choice")
```
